```python
#Re-loading the dataset

from datasets import load_dataset

dataset = load_dataset('SetFit/emotion')
```

```python
#Extracting both the Training and the Testing Data from the re-loaded dataset.

training_data = dataset['train']

test_data = dataset['test']
```

```python
#Tokenization and Model Definition

from transformers import AutoTokenizer

from huggingface_hub import hf_hub_download

import os

import socket

socket.setdefaulttimeout(30)

bert_tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
#Tokenizing both the training and testing data

def tokenize(batch):

  return bert_tokenizer(batch['text'], padding = True, truncation = True)

training_dataset = training_data.map(tokenize, batched = True)

testing_dataset = test_data.map(tokenize, batched = True)
```

```python
#Converting the datasets to PyTorch formats

training_dataset.set_format(type = "torch", columns = ["input_ids", "attention_mask" , "label"])

testing_dataset.set_format(type = "torch", columns = ["input_ids", "attention_mask" , "label"])
```

```python
#Loading the Pretrained LLM Model for Fine-Tuning

from transformers import AutoModelForSequenceClassification

bert_model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels = 6)  #6 Output Classes.
```

```python
print(bert_model)
```

```python
#Initializing the Fine-Tuning Process with Parameter Efficient Fine-Tuning Technique -> Low Rank Adaptation(LoRA)

from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training, TaskType

#Configuring Low Rank Adaptation(LoRA)

lora_config = LoraConfig(

                         r = 8, #Low - Rank Adaptation Dimension

                         lora_alpha = 32, #Scaling Factor

                         lora_dropout = 0.05, #Dropout Rate - To prevent overfitting

                         target_modules = ["query", "values"] #Applying LoRa to only Self-Attention layers

)

#Configuring LoRa with the Pre-trained LLM Model

bert_model = prepare_model_for_kbit_training(bert_model)

#Integrating LoRa into the Pre-Trained LLM Model

lora_model = get_peft_model(bert_model, lora_config)

#Printing the LoRa enabled model

lora_model.print_trainable_parameters()
```

```python
#Initializing the Fine - Tuned Training Process post Fine-Tuning (Applying PEFT Technique - Low Rank Adaptation (LoRa)) on the Pre-trained LLM

from transformers import TrainingArguments, Trainer

#Configuration class that defines how the Training should happen.

training_args = TrainingArguments(

                                        output_dir = "./model_checkpoints",  #Where to save model

                                        num_train_epochs = 3, #Training for 3 epochs

                                        per_device_train_batch_size = 16,  # 16 samples per CPU/GPU

                                        eval_strategy ="epoch", #Evaluation after every epoch

                                        save_strategy = "epoch", #Saving model after every epoch

                                        logging_steps = 10, #Log training metrics every 10 steps

                                        fp16 = True, #Use mixed precision for faster training

                                        save_total_limit = 3,

                                        disable_tqdm = False,

                                        #load_best_model_at_end=True,

                                        #metric_for_best_model = "accuracy",

                                        greater_is_better = True,

                                        report_to = ["wandb"]

)

#A high level class that automates training, evaluation and saving models.
#It wraps around your model and dataset, handling:

# 1 - Training loops

# 2 - Evaluation during training

# 3 - Model saving and checkpointing


lora_model_trainer = Trainer(

                             model = lora_model,

                             args = training_args,

                             train_dataset = training_dataset,

                             eval_dataset = testing_dataset,

                             tokenizer = bert_tokenizer,


)

lora_model_trainer.can_return_loss = True

lora_model_trainer.train()

lora_model_trainer.evaluate()

lora_model_trainer.save_model()
```

```python
#Applying Quantized Low Rank Adaptation(Q-LoRA) Parameter Efficient Fine Tuning Technique

from transformers import BitsAndBytesConfig

import torch

#Applying 4-bit Quantization to Optimize Memory Usage
bnb_config = BitsAndBytesConfig(

                              load_in_4bit = True,  #Enabling 4-bit Quantization

                              bnb_4bit_compute_dtype = torch.float16, #Computing in float16

                              bnb_4bit_use_double_quant = True #Applying Double Quantization for greater efficiency

)

#Configuring Q-LoRA to the Pre-trained LLM Model

bert_model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels = 6, quantization_config = bnb_config)

#Pre-trained LLM Model post applying Quantized Low Rank Adaptation(Q-LoRA) Parameter Efficient Fine Tuning Technique

q_lora_model = get_peft_model(bert_model, lora_config)

#Displaying the Updated Trainable Parameter Details of the Pre-Trained LLM

q_lora_model.print_trainable_parameters()
```
